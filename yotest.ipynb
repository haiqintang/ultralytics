```python
from ultralytics import YOLO
from pathlib import Path

# Create a new YOLO11n-OBB model from scratch
model = YOLO("yolo12n-obb.yaml")
```

```python

# Train the model on the custom dataset
data_yaml = str(Path.home()) + "/samba/private/dataset/yolo_obb/misplace.yaml"
output_folder = str(Path.home()) + "/samba/private/ai_output/yolo12n_obb_misplace"
```

```python

results = model.train(data=data_yaml,
                        epochs=20,
                        device='0',
                        project=output_folder,
                        imgsz=128,
                        batch=128,
                        
                        # === disable space augmentation ===
                        mosaic=0.0,
                        mixup=0.0,
                        copy_paste=0.0,
                        cutmix=0.0,
                        degrees=0.0,
                        translate=0.0,
                        scale=0.0,
                        shear=0.0,
                        perspective=0.0,
                        )
```
